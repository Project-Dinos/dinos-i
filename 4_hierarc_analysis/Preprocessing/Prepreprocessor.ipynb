```python
import pandas as pd
import fitsio
import numpy as np
```

```python
df = pd.read_csv('ext_data/slacs_all_params.csv')
vdisp = fitsio.read('ext_data/vdisp_slacs_asb.fits')
```

```python
gamma_sys_dict = {}

sys_error_df = pd.read_csv('../2_data_tables/tables/DINOS_Lenses_with_systematic_v2.csv')
for index, row in sys_error_df.iterrows():
    gamma_sys_dict[row['Lens Name']] = row['gamma_systematic']
```

```python
z_lens_dict = {}
z_source_dict = {}
v_disp_dict = {}
v_disp_err_dict = {}

for index, row in df.iterrows():
   # print(row['name'], row['z_lens'], row['z_source'])
    z_lens_dict[row['name']] = row['z_lens']
    z_source_dict[row['name']] = row['z_source']
    
for i in vdisp:
    v_disp_dict[i[-1]] = i[-3]
    v_disp_err_dict[i[-1]] = i[-2]
```

```python
mass_df = pd.read_csv('/home/s1/chinyi/80bdata/dinos/final_gold_analysis/2_data_tables/tables/SLACS_Gold_mass_params.csv')
light_df = pd.read_csv('/home/s1/chinyi/80bdata/dinos/final_gold_analysis/2_data_tables/tables/SLACS_Gold_light_params.csv')
result = pd.merge(mass_df, light_df,left_on="Lens_name",right_on="lens_name") 
```

```python
result['name'] = result['Lens_name']
result['theta_E'] = result['R_Ein']
result['theta_E_error'] = 0.5*(result['R_Ein_e1'] + result['R_Ein_e2'])
result['gamma'] = result['γ']
result['gamma_error_uncorr'] = 0.5*(result['γ_e1'] + result['γ_e2'])
result['gamma_error_sys'] = result['name'].map(gamma_sys_dict)
result['gamma_error'] = np.sqrt(result['gamma_error_uncorr']**2 + result['gamma_error_sys']**2)
result['r_eff'] = result['r_eff']
result['r_eff_error_uncorr'] = 0.5*(result['r_eff_e1'] + result['r_eff_e2'])
result['r_eff_error'] = np.sqrt(result['r_eff_error_uncorr']**2 + (0.02*result['r_eff'])**2)
result['z_lens'] = result['name'].map(z_lens_dict)
result['z_source'] = result['name'].map(z_source_dict)
result['sigma_v'] = result['name'].map(v_disp_dict)
result['sigma_v_error'] = result['name'].map(v_disp_err_dict)

result_cut = result[['name','z_lens','z_source','sigma_v','sigma_v_error',
        'theta_E','theta_E_error','gamma','gamma_error','r_eff','r_eff_error','light_model',
        'r_sersic_1','n_sersic_1','amp_sersic_1','r_sersic_2','n_sersic_2','amp_sersic_2'] ]  


####### Hack to get SDSSJ2347-0005 velocity dispersion
result_cut.at[33, 'sigma_v'] = list(df[df['name']=='SDSSJ2347-0005']['sigma_v'])[0]
result_cut.at[33, 'sigma_v_error'] = list(df[df['name']=='SDSSJ2347-0005']['sigma_v_error'])[0]



result_cut.to_csv('SLACS_all_params.csv',index=False)

#name,flag_imaging_tan23,z_lens,z_source,sigma_v,sigma_v_error,theta_E,theta_E_error,gamma,gamma_error,r_eff,r_eff_err
```
