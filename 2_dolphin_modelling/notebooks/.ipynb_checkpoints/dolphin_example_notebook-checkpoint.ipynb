```python
import yaml
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import h5py
import sys
from dolphin.processor import Processor
from dolphin.analysis.output import Output
```

```python
processor = Processor('../')
output = Output('../')
```

```python
def Outputer(lens_name,model_id):

    with open('../settings/{}_config.yml'.format(lens_name)) as file:
        lens_list = yaml.load(file, Loader=yaml.FullLoader)
        
    for i, band in enumerate(lens_list['band']):
        fig = output.plot_model_overview(lens_name=lens_name, model_id=model_id,band_index=i)
        plt.show()
    return None
```

```python
Outputer('SL2SJ1359+5535','dinos_i')
```
